```python
data['text length'] = data['Story_Text'].apply(lambda x: len(x))
```

```python
data.groupby('text length',{'count':sf.aggregate.COUNT}).sort('count',False)
```

```python
data = data[data['text length']>0]
```

```python
data.groupby('Total Expenses',{'count':sf.aggregate.COUNT}).sort('count',False)
```

```python
def condoll(x):
    try: 
        return float(x.replace('$','').replace(' ','').replace(',',''))
    except:
        return None
data['Total Expenses'] = data['Total Expenses'].apply(condoll)
data['Total Revenue'] = data['Total Revenue'].apply(condoll)
```

```python
data['NTEE_Code'] = data['NTEE_Code'].apply(lambda x: x.split(','))
```

```python
data['NTEE_Code'] = data['NTEE_Code'].apply(lambda x: {y.replace(' ',''):1 for y in x})
```
